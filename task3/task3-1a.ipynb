# Schritt 1
Anzahl Messungen $n$ = 50
Wir nehmen an, dass die Messungen unabhängig sind und alle der gleichen Verteilung folgen. $X_1$, $X_2$, ... , $X_n$  haben Erwartungswert µ und Varianz $σ^2$.
Da $n$ hinreichend gross ist, nehmen wir keine konkrete Verteilung für $X_i$ an und argumentieren stattdessen mit dem zentralen Grenzwertsatz für einen annähernd normalverteilten Mittelwert.


# Schritt 2
Aus nachfolgendem Satz und der Aufgabenstellung werden folgende Hypothesen aufgestellt (einseitig):
```Der Müesli-Produzent hat die Vermutung, seine Maschine sei falsch kalibriert und packe zu viel Müesli in die Säcke, was seine Marge drückt.```

$H_0: \mu \leq \mu_0, \qquad H_\text{A}: \mu > \mu_0$

wobei $\mu_0 = 500g$





# Schritt 3

Die Teststatistik $z$ erhalten wir als $\displaystyle z = \frac{\overline{x} - \mu_0}{s/\sqrt{n}} = \frac{501.52-500}{5.54109/\sqrt{50}} \approx 1.939695$.



# Schritt 4
Der zugehörige $p$-Wert ergibt sich als $P(Z \geq 1.939695)= 1- P(Z\leq 1.939695) \approx 0.0262$.

Bzw. pessimistisch aus Normalverteilungstabelle: $0.0268$

# Schritt 5
Da kein Signifikanzniveau angegeben ist, nehmen wir $\alpha = 0.05$.


# Schritt 6
Gemäss Formel:

$
[x_{\text{lower\_limit}}, x_{\text{upper\_limit}}] =
\left[ \bar{x} - z_{0.95} \cdot \frac{s}{\sqrt{n}},\ \infty \right)
$

In $95\%$ der Fällen wird ein Mittelwert gemessen,
der im Interval $[500.231046, \infty)$ liegt.

# Schritt 7
Da $p$-Wert $< \alpha$ ist, wird die Nullhypothese $H_0$ verworfen.
