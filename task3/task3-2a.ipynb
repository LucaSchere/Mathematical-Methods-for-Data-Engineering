# Schritt 1
Anzahl Messungen $n$ = 300
Wir nehmen an, dass die Messungen unabhängig sind und alle der gleichen Verteilung folgen. $X_1$, $X_2$, ... , $X_n$  haben Erwartungswert µ und Varianz $σ^2$.
Da $n$ hinreichend gross ist, nehmen wir keine konkrete Verteilung für $X_i$ an und argumentieren stattdessen mit dem zentralen Grenzwertsatz für einen annähernd normalverteilten Mittelwert.


# Schritt 2
Aus der Aufgabenstellung werden folgende Hypothesen aufgestellt. Da der Investor sicherlich auch bei mehr als 180 CHF im Schnitt einsteigt.

$H_0: \mu \geq \mu_0, \qquad H_\text{A}: \mu < \mu_0$

wobei $\mu_0 = 180 CHF$





# Schritt 3

Die Teststatistik $z$ erhalten wir als $\displaystyle z = \frac{\overline{x} - \mu_0}{s/\sqrt{n}} = \frac{190-180}{75/\sqrt{300}} \approx 2.30940107675850$



# Schritt 4
Der zugehörige $p$-Wert ergibt sich als $P(Z \leq 2.30940) \approx 0.98954$.

Bzw. pessimistisch aus Normalverteilungstabelle: $0.98928$

# Schritt 5
Da kein Signifikanzniveau angegeben ist, nehmen wir $\alpha = 0.05$.


# Schritt 6
Gemäss Formel:

$
[x_{\text{lower\_limit}}, x_{\text{upper\_limit}}] =
\left(-\infty,\ \bar{x} + z_{0.95} \cdot \frac{s}{\sqrt{n}} \right]
$

In $95\%$ der Fällen wird ein Mittelwert gemessen,
der im Interval $[-\infty,197.12)$ liegt.

# Schritt 7
Da $p$-Wert $> \alpha$ ist, wird die Nullhypothese $H_0$ nicht verworfen.
Sprich, der Investor sollte einsteigen.
