# Schritt 1
Anzahl Messungen $n$ = 23.
Das ist eigentlich zu wenig für einen z-Test, da - gemäss zentralem Grenzwertsatz - $n$ mindestens 30 sein sollte, um eine annähernde Normalverteilung des Mittelwerts zu erreichen.


Wir nehmen an, dass die Messungen unabhängig sind und alle der gleichen Verteilung folgen. $X_1$, $X_2$, ... , $X_n$  haben Erwartungswert µ und Varianz $σ^2$.

# Schritt 2
Aus der Aufgabenstellung werden folgende Hypothesen aufgestellt:

$H_0: \mu \leq \mu_0, \qquad H_\text{A}: \mu > \mu_0$

wobei $\mu_0 = 170cm$





# Schritt 3

Die Teststatistik $z$ erhalten wir als $\displaystyle z = \frac{\overline{x} - \mu_0}{s/\sqrt{n}} = \frac{177.3913-170}{9.717552677476244/\sqrt{23}} \approx 3.64777$.



# Schritt 4
Der zugehörige $p$-Wert ergibt sich als $P(Z \geq 3.64777)= 1- P(Z\leq 3.64777) \approx 0.00013$.

Bzw. pessimistisch aus Normalverteilungstabelle: $0.00016$

# Schritt 5
Da kein Signifikanzniveau angegeben ist, nehmen wir $\alpha = 0.05$.


# Schritt 6
Gemäss Formel:

$
[x_{\text{lower\_limit}}, x_{\text{upper\_limit}}] =
\left[ \bar{x} - z_{0.95} \cdot \frac{s}{\sqrt{n}},\ \infty \right)
$

In $95\%$ der Fällen wird ein Mittelwert gemessen,
der im Interval $[174.05842301217247, \infty)$ liegt.

# Schritt 7
Da $p$-Wert $< \alpha$ ist, wird die Nullhypothese $H_0$ verworfen.
